```python
file_path = r'/content/ind_nifty100list.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
fig = px.treemap(df, path=['Industry'], title='Industry Frequency TreeMap')

# Show the TreeMap
fig.show()
```

```python
fig = px.treemap(df, path=['Industry'], title='Industry Frequency TreeMap',
                 color_discrete_sequence=['green', 'blue', 'orange', 'red', 'purple'])

# Customize label appearance
fig.update_layout(treemaplabels=dict(family="Arial", size=12, color='white'))

# Show the TreeMap
fig.show()
```

```python
fig = px.treemap(df, path=['Symbol'], title=' Company Name As Symbol Frequency TreeMap', hover_data=['Company Name'],
                 color_discrete_sequence=px.colors.qualitative.Set1)

# Show the TreeMap
fig.show()
```

```python
fig = px.treemap(df, path=['Industry'], title='Industry Frequency TreeMap',
                 color_discrete_sequence=px.colors.qualitative.Set1)

# Show the TreeMap
fig.show()

# Define a callback function to update the treemap when clicking on a sector
def update_treemap(trace, points, selector):
    selected_sector = points.point_numbers[0] if points.point_numbers else None

    if selected_sector is not None:
        # Filter data based on the selected sector
        selected_sector = df['Industry'].unique()[selected_sector]
        filtered_data = df[df['Industry'] == selected_sector]

        # Create a treemap for the selected sector
        updated_fig = px.treemap(filtered_data, path=['Company Name'],
                                 title=f'Companies in {selected_sector}',
                                 color_discrete_sequence=px.colors.qualitative.Set2)

        # Display the list of companies in the console
        print(f'Companies in {selected_sector}: {filtered_data["Company Name"].tolist()}')

        # Update the layout to show the new treemap
        updated_fig.update_layout(title=f'Companies in {selected_sector}')
        updated_fig.show()

# Add click event to the initial treemap
fig.data[0].on_click(update_treemap)

# Display the initial treemap
fig.show()
```
