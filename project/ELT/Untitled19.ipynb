```python
# Define the directory where the CSV files are stored
directory = 'C:/Users/ASUS/Downloads/Telegram Desktop/New folder/data/dataFile'

# Counter for unique index
index_counter = 1
i=0
# List to keep track of dataframes for final concatenation
dataframes = []

# Process each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory, filename)
        # Read the CSV file
        df = pd.read_csv(file_path)

        # Check if the 'date' column exists in the dataframe
        if 'date' in df.columns:
            # Convert the 'date' column to datetime
            df['datetime'] = pd.to_datetime(df['date'])

            # Split the 'datetime' column into separate 'date' and 'time' columns
            df['date'] = df['datetime'].dt.date
            df['time'] = df['datetime'].dt.time

            # Drop the original 'datetime' column if no longer needed
            df.drop('datetime', axis=1, inplace=True)

        # Add an index column with the current index_counter value
        df['Index'] = index_counter

        # Append the dataframe to the list
        dataframes.append(df)

        # Increment the index_counter
        index_counter += 1
        
        df['Quarter'] = 'Q' + pd.to_datetime(df['date']).dt.quarter.astype(str)        
        new_filename = f"modified_{filename}"  
        print(new_filename)
        print(i)
        dataframes[i].to_csv(new_filename, index=False)
        i+=1

# At this point, each dataframe in 'dataframes' list has the date and time columns split,
# and a new 'Index' column added, but none of them have been saved back to CSV files.
# They can be saved later as needed.
```
